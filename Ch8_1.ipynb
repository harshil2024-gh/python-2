```python
import requests, json
def get_coordinates():
    api_key=input("Enter API key : ")
    city=input("Enter city name : ")
    request_url=f"http://api.openweathermap.org/geo/1.0/direct?q={city}&appid={api_key}"
    response=requests.get(request_url)
    coordinates=response.json()
    print(coordinates)
    print()
    Lattitude=coordinates[0]['lat']
    Longitude=coordinates[0]['lon']
    return Lattitude,Longitude
lat,lon=get_coordinates()
print(lat,lon)
```

```python
import requests, json
def current_weather():
    api_key="e1f152b97c19e6f91585e4c5b3191b78"
    request_url=f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}"
    response=requests.get(request_url)
    data=response.json()
    print(json.dumps(data,indent=5))
    des=data['weather'][0]['description']
    temp=data['main']['temp']
    pres=data['main']['pressure']
    humi=data['main']['humidity']
    w_speed=data['wind']['speed']
    visi=data['visibility']
    return des,temp,pres,humi,w_speed,visi
des,temp,pres,humi,w_speed,visi=current_weather()
print(des,temp,pres,humi,w_speed,visi)
```
