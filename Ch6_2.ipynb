```python
import pandas as pd
df=pd.read_csv("Social_Network_Ads.csv")
df
```

```python
import numpy as np
df['Gender']=np.where(df['Gender']=='Female',0,1)
df
```

```python
x=df.iloc[:,1:4]
y=df['Purchased']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
nn=KNeighborsClassifier(n_neighbors=15)
model=nn.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_pred)
```

```python
diff=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
diff
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
a=accuracy_score(y_test,y_pred)
print(a)
```

```python
import sklearn.metrics
print(sklearn.metrics.recall_score(y_test,y_pred,pos_label=0)) # specificity
print(sklearn.metrics.recall_score(y_test,y_pred)) # Sensitivity
```

```python
#Multiple Linear Regression
import pandas as pd
df=pd.read_csv("Fish.csv")
df
```

```python
df.columns
```

```python
x=df[['Length1', 'Length2', 'Length3', 'Height',
       'Width']]
y=df['Weight']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42) # train_size=0.8
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_pred)
print(type(y_pred))
```

```python
print("Coefficients : ",model.coef_)
print("Intercept : ",model.intercept_)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
print("MSE : ",mean_squared_error(y_test,y_pred))
print("MAE : ",mean_absolute_error(y_test,y_pred))
print("R2 Score : ",r2_score(y_test,y_pred))
```

```python
#Polynomial Regression
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
model_poly=poly.fit_transform(x)
lr=LinearRegression()
lr.fit(model_poly,y)
```

```python
y_pred=lr.predict(model_poly)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
import numpy as np
print("RMSE : ",np.sqrt(mean_squared_error(y,y_pred)))
print("MAE : ",mean_absolute_error(y,y_pred))
print("R2 Score : ",r2_score(y,y_pred))
```
