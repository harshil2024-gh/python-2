# Area plot chart

```python
import matplotlib.pyplot as plt
x=range(1,6)
y=[1,4,6,8,4]
plt.fill_between(x,y,color='skyblue',alpha=0.4)
plt.plot(x,y,color='slateblue',alpha=0.8)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(time,expense,color='red')
plt.fill_between(time,income,expense,where=(income>expense),color='green',alpha=0.25,label="Postive",interpolate=True)
plt.fill_between(time,income,expense,where=(income<=expense),color='red',alpha=0.25,label="Negative",interpolate=True)
plt.xlabel('time')
plt.ylabel('income/expense')
plt.legend()
plt.show()
```

# Box Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data =np.random.rand(100)
print(data)
plt.boxplot(data,widths=0.75,notch=True)
plt.show()
df=pd.DataFrame(data)
df.describe()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x1=5*np.random.rand(40)
x2=5*np.random.rand(40)+25
x3=25*np.random.rand(20)
x=np.concatenate((x1,x2,x3))
y1=5*np.random.rand(40)
y2=5*np.random.rand(40)+25
y3=25*np.random.rand(20)
y=np.concatenate((y1,y2,y3))
color_array=['b']*40+['r']*40+['g']*20
plt.scatter(x,y,s=[100],marker="^",c=color_array)
plt.show()
```

# Waffle Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
data={'phone':['xiaomi','Samsung','Apple','Nokia','Realme'],'stock':[45,12,8,5,50]}
df=pd.DataFrame(data)
plt.figure(FigureClass=Waffle,rows=10,values=df.stock,labels=list(df.phone))
plt.show()
```

# Word Cloud

```python
from wordcloud import WordCloud,STOPWORDS
alice_novel=open('alice_in_wonderland.txt')
alice_novel=alice_novel.read()
stopwords=set(STOPWORDS)
print(stopwords)
print(len(stopwords))
```

```python
import matplotlib.pyplot as plt
alice_wc=WordCloud(background_color='white',max_words=2000,stopwords=stopwords)
alice_wc.generate(alice_novel)
plt.imshow(alice_wc)
plt.show()
```
