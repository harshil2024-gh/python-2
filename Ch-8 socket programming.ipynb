## socket programming
- TCP-> transmission control protocol (type=socket.SOCK_STREAM)->default socket type
- UPD-> user Datagram protocol

```python
%%writefile server_tcp.py
import socket
host=socket.gethostname()
port=2000
Addr=(host,port)
server_socket=socket.socket()
server_socket.bind(Addr)
server_socket.listen()
conn,addr=server_socket.accept()
print("Connection from ",str(addr))
while True:
    data=conn.recv(1024).decode()
    if not data:
        break
#     print(data)
    data=data.upper()
    conn.send(data.encode())
conn.close()
```

```python
%%writefile client_tcp.py
import socket
host='192.168.104.25'
port=2000
# Addr=(host,port)
client_socket=socket.socket()
client_socket.connect((host,port))
message=input('->')
while message!="":
    client_socket.send(message.encode())
    data=client_socket.recv(1024).decode()
    print("Recieved from server->",data)
    message=input('->')
client_socket.close()
```

### udp ->connectionless

```python
%%writefile udp_server.py
import socket
host=""
port=5000
udp_server=socket.socket(type=socket.SOCK_DGRAM)
udp_server.bind((host,port))
while True:
    print("Waiting for message")
    data,addr=udp_server.recvfrom(1024)
    print("Recieved:",data.decode(),"from",addr)
    msg=input("Enter Message:")
    udp_server.sendto(msg.encode(),addr)
udp_server.close()
```

```python
#Web scrapping of multiple pages
import requests
from bs4 import BeautifulSoup
upperframe=[]
for i in range(1,5):
    req_url="https://www.politifact.com/factchecks/?page="+str(i)
    result=requests.get(req_url)
    content=result.text
    soup=BeautifulSoup(content,'html.parser')
    frame=[]
    links=soup.find_all('li',class_='o-listicle__item')
    for j in links:
        statement=j.find("div",class_="m-statement__quote").text.strip()
        link="https://politifact.com"
        link+=j.find("div",class_="m-statement__quote").find("a")['href'].strip()
        Date=j.find('footer',class_="m-statement__footer").text.split("•")[1].strip()
        
        Source=j.find("a",class_='m-statement__name').text.strip()
        frame.append((statement,link,Date,Source))
    upperframe.extend(frame)
```
