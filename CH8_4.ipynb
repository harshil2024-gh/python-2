```python
import requests
from bs4 import BeautifulSoup
request_url="https://subslikescript.com/movie/Lego_Star_Wars_The_Padawan_Menace-2005268"
result=requests.get(request_url)
content=result.text
```

```python
soup=BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
transcript=soup.find('div',class_='full-script').text.strip()
print(transcript)
```

```python
with open("Lego_Star_Wars.txt", "w",encoding='utf-8') as file:
    file.write(transcript)
```

```python
import requests
from bs4 import BeautifulSoup
request_url="https://subslikescript.com/movies"
result=requests.get(request_url)
content=result.text
```

```python
import requests
from bs4 import BeautifulSoup
request_url="https://www.politifact.com/factchecks/"
result=requests.get(request_url)
content=result.text
```

```python
soup=BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
x=soup.find_all('div',class_="m-statement__quote")
print(x)
```

```python
l=[]
for i in x:
    y=i.find('a')
    l.append(y.text.strip())
print(l)
print(len(l))
```

```python
z=soup.find_all('footer',class_="m-statement__footer")
# print(z)
d=[]
for i in z:
    d.append(i.text.split('•')[1].strip())
print(d)
```

```python
w=soup.find_all('div',class_="m-statement__desc")
e=[]
for i in w:
    e.append(i.text.split(',')[1].strip())
print(e)
```

```python
import pandas as pd
df=pd.DataFrame({"Quotes":l,"Date":d,"Source":e})
df
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(open('top250.html','r',encoding="utf-8"))
print(soup.prettify())
```

```python
x=soup.find_all('td',class_="titleColumn")
l=[]
year=[]
for i in x:
    y=i.find('a')
    l.append(y.text.strip())
    c=i.find('span')
    year.append(c.text[1:-1].strip())
print(len(l))
print(len(year))
```

```python
r=soup.find_all('td',class_="ratingColumn imdbRating")
rating=[]
for i in r:
    b=i.find('strong')
    rating.append(b.text.strip())
print(len(rating))
```

```python
import pandas as pd
df=pd.DataFrame({"Title":l,"Rating":rating,"Year":year})
df
```
