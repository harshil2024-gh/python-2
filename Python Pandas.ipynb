### Python Series: 
- Series is like a column in a table.
- 1-D array holding data of any type.

```python
import pandas as pd
a=[1,7,2]
myvar=pd.Series(a)
print(myvar)
print(myvar[1])
print(type(myvar))
```

```python
import pandas as pd
a=[1,7,2]
myvar=pd.Series(a,index=["x","y","z"])
print(myvar)
print(myvar["x"])
print(myvar[1])
```

# DataFrame
- 2 dimensional data stucture like 2D array or a table with rows and column.

```python
import pandas as pd
data={"calories":[420,300,390],
     "duration":[50,40,45]}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.loc[0])
print()
print(dict(df.loc[0]))
print(list(df.loc[0]))
```

```python
#iloc-->integer location
print(df.iloc[0])
```

```python
print(df['calories']) #or print(df.calories)
print(df['duration']) #or print(df.duration)
```

```python
print(df.loc[[0,1]])
```

```python
df=pd.DataFrame(data,index=['day1','day2','day3'])
print(df)
```

```python
print(df.loc['day1'])
# print(df.iloc['day1']) -->Error. ->Only integer index is allowed
```

```python
print(df.iloc[0])
```

```python
# print(df.loc[0:2]) -->Error. because indexs are: ['day1','day2','day3']
```

```python
print(df.iloc[0:2])
```

```python
import pandas as pd
data={"calories":[420,300,390],
     "duration":[50,40,45]}
df=pd.DataFrame(data)
print(df)
print()
print(df.iloc[0:2]) #In  iloc[0:2] 2 is excluded
print()
print(df.loc[0:2])#In  loc[0:2] 2 is included
```

# Parallel coordinates

```python
import pandas as pd
dataset=pd.read_csv('auto-mpg.csv')
```

```python
from pandas.plotting import parallel_coordinates
pll=parallel_coordinates(dataset,'cylinders',cols=["acceleration","mpg"],color=["red","blue","green","yellow","orange"])
```

```python
pll=parallel_coordinates(dataset,'displacement',cols=["acceleration","mpg"])
```

```python
pll=parallel_coordinates(dataset,'cylinders',cols=["cylinders","mpg","origin"],color=["red","blue","green","yellow","orange"])
```
